```python
#Q1
import matplotlib.pyplot as plt
year= [2010, 2020, 2030, 2040, 2050, 2060, 2070, 2080, 2090, 2100]
pop = ["5B", "10B", "15B", "20B", "25B", "30B", "35B", "40B", "45B", "50B"]
plt.title("Population Inflamation with Time")
plt.xlabel("Years")
plt.ylabel("Population")
plt.plot(year, pop)
plt.show()
print("Population of World in Year:", year[-1],"will be", pop[-1],"(Billion)")
```

```python
#Q3
import matplotlib.pyplot as plt
import pandas as pd 
df=pd.read_csv('http://assets.datacamp.com/course/intermediate_python/gapminder.csv', index_col = 0)
print(df)
gdp_cap = list(df.gdp_cap)
life_exp = list(df.life_exp)

print("Last Item GDP (Zimbabwe):", gdp_cap[-1], "Last Item Life_Exp (Zimbabwe):", life_exp[-1])

plt.xlabel("GDP Cap")
plt.ylabel("Life Exp")
plt.title("Proffessor Hans Rosling Research")
plt.plot(gdp_cap, life_exp)
plt.show()
plt.clf()
```
