```python

dataset = pd.read_csv('Data_1.csv')

x = dataset['SAT']
y= dataset['GPA']

mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(1600, 2000, 10)


plt.scatter(x, y)
plt.plot(myline, mymodel(myline))

print(r2_score(y, mymodel(x)))

print('Predicted Value', mymodel(2100))
```

```python
dataset = pd.read_csv('Data_1.csv')
data_scaled = normalize(dataset)
data_scaled = pd.DataFrame(data_scaled, columns=dataset.columns)
plt.figure(figsize=(12, 8))  
plt.title("Dendrograms")  
#here we made the Dendrograms
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))

#here we defined the Threshold Line
plt.axhline(y=0.0005, color='r', linestyle='--')

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(dataset)
```
