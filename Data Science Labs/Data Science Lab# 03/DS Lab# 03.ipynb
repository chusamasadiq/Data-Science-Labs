```python
import pandas as pd
import matplotlib.pyplot as plt
dataset= pd.read_csv("weather.csv")
print(dataset)


plt.plot(dataset['outlook'], dataset['humidity'] )
plt.show()

plt.scatter(dataset['outlook'], dataset['humidity'] )
plt.show()

plt.bar(dataset['outlook'], dataset['humidity'] )
plt.show()

plt.pie(dataset['outlook'])
plt.show()


plt.hist(dataset['outlook'], bins=4 )
plt.show()

data_1 = pd.DataFrame(dataset['outlook'], dataset['windy'])
print(data_1)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
dataset= pd.read_csv("weather.csv")
print(dataset)

data_1 = pd.DataFrame(dataset['outlook'], dataset['windy'])
print(data_1)
sns.lmplot(x='outlook', y='windy', data=dataset, fit_reg=False, hue='play')
sns.boxplot(data=dataset)
```
